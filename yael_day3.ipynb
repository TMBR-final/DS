```python
df = pd.read_csv('pupils.csv')
```

```python
df.drop(['Name','Country','gen'],axis=1,inplace=True)
```

```python
df['incp']=df.income/df.family
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['Age', 'Weight','type','incp']]
```

```python
y = df.avg
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33,random_state=101)
```

```python
import sklearn.linear_model as slin
```

```python
#model = slin.LinearRegression()
model = slin.ARDRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
y_pred = model.predict(X_test)
```

```python
(y_pred - y_test).abs().mean()
```
